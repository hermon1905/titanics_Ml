### Encodons les caracteristiques categorielles

```python
categorical = train_df.select_dtypes(include = "object").columns
le = LabelEncoder()

for col in categorical :
    train_df[col] = le.fit_transform(train_df[col])

categorical1 = test_df.select_dtypes(include = "object").columns
le_ = LabelEncoder()
for col in categorical1 :
    test_df[col] = le.fit_transform(test_df[col])
```

### **appliquons notre modele**

```python
X = train_df.drop(['Survived', 'PassengerId'], axis=1)
y = train_df['Survived']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
```

```python
y_pred
```

```python
np.array(y_val)
```

```python
accuracy_score(y_val, y_pred)
```

```python
print(classification_report(y_val, y_pred))
```
